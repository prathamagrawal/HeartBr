```python
import pandas as pd
import numpy as np 
import missingno as msno
```

```python
dataraw=pd.read_csv('framingham.csv')
```

```python
dataraw
```

```python
dataraw=dataraw.dropna(0,subset=['education','glucose','BPMeds','heartRate'])
```

```python
dataraw
```

```python
dataraw['cigsPerDay']=dataraw['cigsPerDay'].replace(np.NaN,dataraw['cigsPerDay'].mean())
dataraw['BMI']=dataraw['BMI'].replace(np.NaN,dataraw['BMI'].mean())
dataraw['totChol']=dataraw['totChol'].replace(np.NaN,dataraw['totChol'].mean())
```

```python
dataraw.to_csv('dataset.csv',index=False)
```
